## 경기도Map ##

```python
kyeongki_map = folium.Map(location=[37.310813, 127.196330],
                 zoom_start=9)
kyeongki_map

#경기도중앙
```

## 19'경기도 노인사고다발지역 시각화 ##

```python
# feature_kyk_df.reset_index(inplace=True)
for i in range(len(kyk_df)):
    marker01 = folium.Circle([kyk_df.loc[i]['X'], kyk_df.loc[i]['Y']],
                        radius = 150,
                        color='red',
                        fill_color='red',
                        tooltip = kyk_df.loc[i]['입력주소'])
#                         icon = folium.Icon(color='red'))
    marker01.add_to(kyeongki_map)
    
    
    
    
#19노인사고반복문
```

```python
kyeongki_map
```

## 노인사고다발지역(red) VS 노인보호구역 시각화(blue)##

```python
for i in range(len(kyk_df2)):
    marker02 = folium.Circle([kyk_df2.loc[i]['위도'], kyk_df2.loc[i]['경도']], #X,Y바뀜
                        radius = 150,
                        color='blue',
                        fill_color='blue',
                        tooltip = kyk_df2.loc[i]['소재지도로명주소'])
    marker02.add_to(kyeongki_map)
    
```
